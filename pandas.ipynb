```python
data = {
    "Name": ["Amit","Riya","Rahul","Sneha","Arjun","Priya","Karan","Neha"],
    "Age": [20,21,22,21,23,22,20,21],
    "Department": ["CS","CS","IT","IT","CS","IT","CS","IT"],
    "Marks": [85,78,92,74,88,95,67,80],
    "StudyHours": [3,2,4,2,5,4,1,3]
}

df = pd.DataFrame(data)

df
```
