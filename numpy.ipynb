```python
#creating a numpy 1-d array with np alias
arr1 = np.array([1, 2, 3, 4, 5])

print(arr1)

#typre of numpy array
print(type(arr1))
```

```python
#2-d array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])

print(arr2)
```

```python
#3-d array
arr3 = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])

print(arr3)
```

```python
print(f'dimension of arr1: {arr1.ndim}d')
print(f'dimension of arr1: {arr2.ndim}d')
print(f'dimension of arr1: {arr3.ndim}d')
```
